```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read data
data_path = r'D:\Desktop\Study in France\CRSS of TI\CRSS data of all containing 13 features 6.11.xlsx'
df = pd.read_excel(data_path)
```

```python
# One-hot encoding for non-numeric feature
df = pd.get_dummies(df, columns=[df.columns[6]])
df
```

```python
plt.figure(figsize=(12, 5))
corrmat = df.corr()
sns.heatmap(corrmat, annot=True, cmap='coolwarm', fmt=".2f")

plt.savefig(r'D:\Desktop\Study in France\CRSS of TI\juypter running results\pictures\Correlation coefficient of 13 features (original) 6.24.png',
            dpi=300,
            bbox_inches='tight')

plt.show()
plt.close()
```

```python
plt.figure(figsize=(8, 6))


plt.figure(figsize=(8, 6))
plt.hist(df.iloc[:, 2], bins=30, density=True, alpha=0.5, color='b')
sns.kdeplot(df.iloc[:, 2], color='r')
plt.xlim(0, 2300)
    
# 设置x轴和y轴标签字号大小
plt.xlabel('CRSS of Π1<a>', fontsize=18)
plt.ylabel('Density', fontsize=18)
    
# 设置x轴和y轴数字的大小
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
    
plt.savefig(r'D:\Desktop\Study in France\CRSS of TI\juypter running results\pictures\Histograms_of_Π1a.png',
            dpi=300,
            bbox_inches='tight')

plt.show()
plt.close()
```
